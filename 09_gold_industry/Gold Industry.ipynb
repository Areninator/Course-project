### Проверка расчета эффективности обогащения

Проверим расчет эффективности 'rougher.output.recovery' с помощю функции Recovery

где:

* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']
```

```python
recovery = ((C * (F - T)) / (F * (C - T))) * 100
```

### Сравние распределения размеров гранул исходного сырья на обучающей и тестовой выборках.

Сформулируем гипотезы:
* Нулевая гипотеза - средние значения размеров гранул исходного сырья на обучающей и тестовой выборках равны. 
* Альтернативная гипотеза - средние значения размеров гранул исходного сырья на обучающей и тестовой выборках различны. 

Если p-value меньше уровня значимости alpha, то нулевая гипотеза отвергается в пользу альтернативной.

```python
alpha = 0.05

results = ttest_ind(data_train['rougher.input.feed_size'], data_test['rougher.input.feed_size'], equal_var=False)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу. Средние значения различны.')
else:
    print('Не получилось отвергнуть нулевую гипотезу. Средние значения равны.')
```

При уровне значимости alpha = 0.05 мы не можем утверждать, что средние значения размеров гранул исходного сырья на обучающей и тестовой выборках различаются.

```python
size = pd.concat([data_train['rougher.input.feed_size'], data_test['rougher.input.feed_size']], axis = 1)
size.columns = ['Обучающая выборка', 'Тестовая выборка']
bins = int(max(size['Обучающая выборка']) - min(size['Обучающая выборка']) + 1)
sns.histplot(data=size,bins=bins,kde=True,alpha = .4,stat='density', common_norm=False)
plt.xlim(0,150);
plt.title('Распределения плотности гранул сырья на обучающей и тестовой выборках', fontsize = 15,loc='center', pad=30)
plt.show();
```

### Функция для вычисления итоговой sMAPE.


Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.
Метрика sMAPE вычисляется так:

```python
def smape(true, predict):
    return 1 / len(true) * sum( abs(true - predict) / ( (abs(true) + abs(predict)) * 0.5) ) * 100
```

Итоговая метрика складывается из двух величин:

```python
def final_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```
